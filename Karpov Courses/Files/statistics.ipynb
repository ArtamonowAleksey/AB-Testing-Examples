```python
import pandas as pd
import random
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import simps
import warnings


%matplotlib inline
```

```python
warnings.filterwarnings('ignore')
```

```python

''' Вероятность того что наше значение будет лежать в заданном интервале '''

def probability(interval,start,stop):
    
    kde = sns.kdeplot(interval)
    x = kde.lines[0].get_xdata()
    y = kde.lines[0].get_ydata()
    mask = (x>=start) & (x<=stop)
    x_in = x[mask]
    y_in = y[mask]
    area = round(simps(y_in,x_in),3)
    
    sns.kdeplot(interval)
    
    return print(f'Вероятность того что наше значение лежит в интервале от {start} до {stop} = {area*100}%')
```

```python
rand_items_normal = np.random.normal(loc=0,scale=1,size=1000)

probability(rand_items_normal,2,3)
```
